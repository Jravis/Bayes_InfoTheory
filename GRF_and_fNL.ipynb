# Gaussian random fields and local non-Gaussianity

```python
import numpy as np
import scipy.linalg
from scipy.stats import norm
from matplotlib import pyplot as plt
%matplotlib inline
```

## Generate white noise

```python
n=1024
rv=norm(loc=0.,scale=1.)
x=np.linspace(-3.,3,100)
normalsim=np.random.normal(0.,1.,n)
plt.xlim(0,n)
plt.scatter(np.arange(n),normalsim,marker='.',color='black')
```

```python
plt.plot(x, rv.pdf(x), color='black', lw=2)
plt.hist(normalsim, bins='auto', normed=True, color='black', histtype='step', lw=1.5)
plt.show()
```

## Setup covariance and generate Gaussian signal

### Covariance demo 1

```python
covar=np.array([[np.exp(-np.abs((i-j)/20.)) for i in xrange(n)] for j in xrange(n)])
plt.matshow(covar, cmap='RdBu')
```

```python
plt.xlim(0,n)
plt.plot(np.arange(n),covar[n/2])
```

```python
rootcovar=scipy.linalg.sqrtm(covar)
signal=rootcovar.dot(normalsim).real
plt.xlim(0,n)
plt.scatter(np.arange(n),signal,marker='.',color='blue')
```

```python
plt.plot(x, rv.pdf(x), color='black', lw=2)
plt.hist(signal, bins='auto', normed=True, histtype='step', lw=1.5)
plt.show()
```

### Covariance demo 2

```python
covar=np.array([[(1. if i<n/2 else 100.) if i==j else 0. for i in xrange(n)] for j in xrange(n)])
plt.matshow(covar, cmap='RdBu')
```

```python
plt.xlim(0,n)
plt.plot(np.arange(n),covar[n/2])
```

```python
rootcovar=scipy.linalg.sqrtm(covar)
signal=rootcovar.dot(normalsim).real
plt.xlim(0,n)
plt.scatter(np.arange(n),signal,marker='.',color='blue')
```

```python
plt.plot(x, rv.pdf(x), color='black', lw=2)
plt.hist(signal, bins='auto', normed=True, histtype='step', lw=1.5)
plt.show()
```

### Covariance demo 3

```python
covar=np.array([[1./(1+np.abs((i-j)/2.))**2 for i in xrange(n)] for j in xrange(n)])
plt.matshow(covar, cmap='RdBu')
```

```python
plt.xlim(0,n)
plt.plot(np.arange(n),covar[n/2])
```

```python
rootcovar=scipy.linalg.sqrtm(covar)
signal=rootcovar.dot(normalsim).real
plt.xlim(0,n)
plt.scatter(np.arange(n),signal,marker='.',color='blue')
```

```python
plt.plot(x, rv.pdf(x), color='black', lw=2)
plt.hist(signal, bins='auto', normed=True, histtype='step', lw=1.5)
plt.show()
```

### Covariance demo 4

```python
covar=np.array([[np.cos((i-j)/8.) for i in xrange(n)] for j in xrange(n)])
plt.matshow(covar, cmap='RdBu')
```

```python
plt.xlim(0,n)
plt.plot(np.arange(n),covar[n/2])
```

```python
rootcovar=scipy.linalg.sqrtm(covar)
signal=rootcovar.dot(normalsim).real
plt.xlim(0,n)
plt.scatter(np.arange(n),signal,marker='.',color='blue')
```

## Generate non-Gaussian signal

```python
fNL=0.2
```

```python
NGsignal=normalsim+fNL*normalsim**2
plt.xlim(0,n)
plt.scatter(np.arange(n),NGsignal,marker='.',color='purple')
```

```python
plt.hist(NGsignal, bins='auto', normed=True, histtype='step', color='purple', lw=1.5)
plt.plot(x, rv.pdf(x), color='black', lw=2)
plt.show()
```
